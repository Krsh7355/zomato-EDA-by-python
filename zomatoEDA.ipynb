# zomata data analysis project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\km113\Downloads\Zomato data .csv")
```

```python
df
```

# datacleaning - convert the data type of column - rate

```python
def handlerate(value):
    value=str(value).split("/")
    value=value[0]
    return float(value)

df["rate"]=df["rate"].apply(handlerate)
print(df.head())
```

```python
df
```

```python
df.info()
```

# question 1 type of restaurant

```python
df.head()
```

```python
sns.countplot(x=df["listed_in(type)"])
plt.xlabel("type of restaurant")
plt.show()
```

# conclusion- majority of the restaurant falls in dining category

```python
df.head()
```

```python
# uestion 2
group_data=df.groupby("listed_in(type)")["votes"].sum()
result=pd.DataFrame({"votes":group_data})
plt.plot(result,c="brown",marker="o")
plt.xlabel("type of restaurant",c="red")
plt.ylabel("vote",c="red")
```

conclusion - dining restaurant recieved maximum votes

# question 4 - which mode recieve maximin rating

```python
df.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x="online_order",y="rate",data=df)	
plt.show()
```

# question 5 which type of restaurant recieve most order

```python
df.head()
```

```python
pivot_table= df.pivot_table(index="listed_in(type)",columns="online_order",aggfunc="size",fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt="d")
plt.title("Heatmap")
plt.xlabel("online_order")
plt.ylabel("listed_in(type)")
plt.show()
```
